```python
import numpy as np
import cv2
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D
from keras.optimizers import Adam
from keras.layers import MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_dir = 'train'
val_dir = 'test'
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(48,48),
        batch_size=64,
        color_mode="grayscale",
        class_mode='categorical')

validation_generator = val_datagen.flow_from_directory(
        val_dir,
        target_size=(48,48),
        batch_size=64,
        color_mode="grayscale",
        class_mode='categorical')
```

```python
emotion_model = Sequential()
emotion_model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(48,48,1)))
emotion_model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))
emotion_model.add(Dropout(0.25))
emotion_model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))
emotion_model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
emotion_model.add(MaxPooling2D(pool_size=(2, 2)))
emotion_model.add(Dropout(0.25))
emotion_model.add(Flatten())
emotion_model.add(Dense(1024, activation='relu'))
emotion_model.add(Dropout(0.5))
emotion_model.add(Dense(7, activation='softmax'))
```

```python
from keras.optimizers import Adam

emotion_model.compile(loss='categorical_crossentropy', optimizer=Adam(learning_rate=0.0001), metrics=['accuracy'])

```

```python
import os
import cv2
import numpy as np
import math
def extract_frames(video_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video_name = os.path.basename(video_path)
    video_name_without_extension = os.path.splitext(video_name)[0]

    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    for frame_num in range(frame_count):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)

        ret, frame = cap.read()

        if not ret:
            break

        x = frame_num % math.floor(frame_width / 200) * 200
        y = frame_num % math.floor(frame_height / 200) * 200

        cropped_frame = frame[y:y+200, x:x+200]

        output_image_path = os.path.join(output_folder, f"{video_name_without_extension}frame{frame_num}.jpg")
        cv2.imwrite(output_image_path, cropped_frame)

video_path='/content/runs/detect/predict/ABA Therapy - Play.avi'
output_folder='/content/extracted_frames'
extract_frames(video_path, output_folder)
```

```python
import cv2
from keras.models import load_model
import numpy as np
import os

# Load the emotion recognition model
emotion_model = load_model('model.h5')

# Function to perform emotion recognition on a frame and add the emotion label
def add_emotion_label(frame):
    # Preprocess frame for emotion recognition
    face = cv2.resize(frame, (48, 48))
    face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
    face = face / 255.0
    face = np.expand_dims(face, axis=0)
    face = np.expand_dims(face, axis=-1)

    # Predict emotion using the emotion recognition model
    emotion_prediction = emotion_model.predict(face)
    emotion_label = np.argmax(emotion_prediction)

    # Add emotion label text to the frame
    emotion_text = "Emotion: " + str(emotion_label)
    cv2.putText(frame, emotion_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    return frame

# Function to process each frame of the video, add emotion labels, and write to a new video file
def create_video_with_emotion_labels(input_video_path, output_video_path):
    cap = cv2.VideoCapture(input_video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Add emotion label to the frame
        frame_with_label = add_emotion_label(frame)

        # Write the frame with emotion label to the output video
        out.write(frame_with_label)

    # Release the VideoCapture and VideoWriter objects
    cap.release()
    out.release()

# Example usage
input_video_path = '/content/runs/detect/predict/ABA Therapy - Play.avi'
output_video_path = '/content/annotated_video_with_emotion_labels.avi'
create_video_with_emotion_labels(input_video_path, output_video_path)
```

```python
import cv2
import numpy as np
from keras.models import load_model

# Load YOLOv8 model for object detection
net = cv2.dnn.readNet('yolov8m.pt', 'yolov8m.cfg')
classes = []
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]
layer_names = net.getLayerNames()
output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]

# Load the emotion recognition model
emotion_model = load_model('model.h5')

# Define a dictionary mapping emotion labels to emotion names
emotion_names = {
    0: "Angry",
    1: "Disgust",
    2: "Fear",
    3: "Happy",
    4: "Sad",
    5: "Surprise",
    6: "Neutral"
}

# Function to detect faces in a frame using YOLOv8
def detect_faces(frame):
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    detected_faces = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5 and class_id == 0:  # Assuming class_id 0 corresponds to face
                center_x = int(detection[0] * frame.shape[1])
                center_y = int(detection[1] * frame.shape[0])
                w = int(detection[2] * frame.shape[1])
                h = int(detection[3] * frame.shape[0])
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                detected_faces.append({'x': x, 'y': y, 'w': w, 'h': h})
    return detected_faces

# Function to detect objects in a frame using YOLOv8
def detect_objects(frame):
    blob = cv2.dnn.blobFromImage(frame, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(output_layers)

    objects = []
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5:
                objects.append({
                    'class_id': class_id,
                    'confidence': confidence,
                    'box': detection[0:4]
                })

    return objects

# Function to analyze human-object interactions
def analyze_interactions(faces, objects):
    interactions = []
    for face in faces:
        for obj in objects:
            # Calculate the distance between the center of the face and the object
            face_center = (face['x'] + face['w'] / 2, face['y'] + face['h'] / 2)
            obj_center = (obj['box'][0] * frame_width + obj['box'][2] * frame_width) / 2, \
                         (obj['box'][1] * frame_height + obj['box'][3] * frame_height) / 2
            distance = np.linalg.norm(np.array(face_center) - np.array(obj_center))

            # Assuming an interaction if the object is within a certain distance from the face
            if distance < 100:  # Adjust this threshold as needed
                interactions.append({
                    'face': face,
                    'object': obj,
                    'distance': distance
                })
    return interactions

# Function to add interaction labels to the frame
def add_interaction_labels(frame, interactions):
    for interaction in interactions:
        face_x, face_y, face_w, face_h = interaction['face']['x'], interaction['face']['y'], \
                                         interaction['face']['w'], interaction['face']['h']
        obj_label = classes[interaction['object']['class_id']]
        cv2.putText(frame, f"{obj_label} interaction", (face_x, face_y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, (0, 255, 0), 2)
    return frame

# Function to process each frame of the video and analyze human-object interactions
def analyze_video(input_video_path, output_video_path):
    cap = cv2.VideoCapture(input_video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Detect faces in the frame
        faces = detect_faces(frame)

        # Detect objects in the frame
        objects = detect_objects(frame)

        # Analyze human-object interactions
        interactions = analyze_interactions(faces, objects)

        # Add interaction labels to the frame
        frame_with_labels = add_interaction_labels(frame, interactions)

        # Write the frame with interaction labels to the output video
        out.write(frame_with_labels)

    # Release the VideoCapture and VideoWriter objects
    cap.release()
    out.release()

# Example usage
input_video_path = '/content/ABA Therapy - Play.mp4'
output_video_path = 'output_video_with_interactions.avi'
analyze_video(input_video_path, output_video_path)
```

```python
import cv2
import numpy as np
import os
import json

# Load the emotion recognition model
emotion_model = load_model('/content/model.h5')

# Define a dictionary mapping emotion labels to emotion names
emotion_names = {
    0: "Angry",
    1: "Disgust",
    2: "Fear",
    3: "Happy",
    4: "Sad",
    5: "Surprise",
    6: "Neutral"
}
import cv2

def get_faces(frame):
    # Load the pre-trained Haar cascade for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Convert the frame to grayscale (Haar cascades work with grayscale images)
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame using the Haar cascade
    faces = face_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Format the detected faces as a list of dictionaries
    detected_faces = [{'x': x, 'y': y, 'w': w, 'h': h} for (x, y, w, h) in faces]

    return detected_faces

def read_detected_objects(folder_path):
    detected_objects = {}
    for filename in os.listdir(folder_path):
        # Split the filename into parts using '.' as a delimiter
        filename_parts = filename.split('.')

        # Check if the filename has the expected format (contains frame number)
        if len(filename_parts) != 2 or not filename_parts[0].isdigit():
            # Skip files that do not match the expected format
            continue

        # Extract frame number from the filename
        frame_number = int(filename_parts[0])

        # Read the detected objects data from the file
        with open(os.path.join(folder_path, filename), 'r') as file:
            data = json.load(file)
            detected_objects[frame_number] = data

    return detected_objects

# Function to analyze human-object interactions
def analyze_interactions(faces, objects):
    interactions = []
    for face in faces:
        for obj in objects:
            # Calculate the distance between the center of the face and the object
            face_center = (face['x'] + face['w'] / 2, face['y'] + face['h'] / 2)
            obj_center = (obj['box'][0] + obj['box'][2] / 2, obj['box'][1] + obj['box'][3] / 2)
            distance = np.linalg.norm(np.array(face_center) - np.array(obj_center))

            # Assuming an interaction if the object is within a certain distance from the face
            if distance < 100:  # Adjust this threshold as needed
                interactions.append({
                    'face': face,
                    'object': obj,
                    'distance': distance
                })
    return interactions

# Function to add interaction labels to the frame
def add_interaction_labels(frame, interactions):
    for interaction in interactions:
        face_x, face_y, face_w, face_h = interaction['face']['x'], interaction['face']['y'], \
                                         interaction['face']['w'], interaction['face']['h']
        obj_label = interaction['object']['label']
        cv2.putText(frame, f"{obj_label} interaction", (face_x, face_y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, (0, 255, 0), 2)
    return frame

# Function to process each frame of the video and analyze human-object interactions
def analyze_video(input_video_path, detected_objects_folder, output_video_path):
    cap = cv2.VideoCapture(input_video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    # Read pre-detected objects data
    detected_objects = read_detected_objects(detected_objects_folder)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES))

        # Get faces from face detection (modify this part according to your implementation)
        faces = get_faces(frame)

        # Get objects detected in the current frame
        objects = detected_objects.get(frame_number, [])

        # Analyze human-object interactions
        interactions = analyze_interactions(faces, objects)

        # Add interaction labels to the frame
        frame_with_labels = add_interaction_labels(frame, interactions)

        # Write the frame with interaction labels to the output video
        out.write(frame_with_labels)

    # Release the VideoCapture and VideoWriter objects
    cap.release()
    out.release()

# Example usage
input_video_path = '/content/ABA Therapy - Play.mp4'
detected_objects_folder = '/content/extracted_frames'
output_video_path = '/content/output_video_with_interactions.avi'
analyze_video(input_video_path, detected_objects_folder, output_video_path)
```

```python
 from google.colab.patches import cv2_imshow
import cv2 as cv
import numpy as np

class PupilDetector:
    def __init__(self) -> None:
        pass

    def detect(self, image_matrix, threshold_correction = 10):
        # the original image matrix in BGR color mode
        self.image_matrix = image_matrix

        # coverting the BGR image matrix to gray
        self.gray_matrix = cv.cvtColor(self.image_matrix, cv.COLOR_BGR2GRAY)
        self.threshold_correction = threshold_correction

        # threshold value for
        threshold = self.getMinThreshold()

        # thresholding the gray image
        thresh = self.thresholding(threshold = threshold)

        # finding the contours in the thresholded image
        contours = self.getContours(thresh)

        # the center cordinate and the radius of pupil enclosing minimum circle
        x, y, radius = self.findPupil(contours)
        return (x, y, radius), self.image_matrix

    def getMinThreshold(self):
        # the pupil is suppose to be the darkest region in the gray image
        # finding the darkest cell intensity and adding a threshold correction
        # for error in colour gradient within the pupil
        min_threshold = 255
        for i in range(len(self.gray_matrix)):
            for j in range(len(self.gray_matrix[0])):
                min_threshold = min(min_threshold, self.gray_matrix[i][j])
        return min_threshold + self.threshold_correction

    def thresholding(self, threshold = 127, maxValue = 255):
        ret, thresh = cv.threshold(self.gray_matrix, threshold, maxValue, cv.THRESH_BINARY)
        return thresh

    def getContours(self, thresh):
        contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
        return contours
    def findPupil(self, contours):
        if not contours:  # Check if contours list is empty
            return None, None, None  # Return None values if no contours are found

    # Finding the max and the second max contour in terms of area
    # The second max is required as the pupil is supposed to be the second max contour
    # The First max is the boundary of the image (OpenCV property)
        A1 = cv.contourArea(contours[0])
        max_area = [A1, contours[0]]

        for cnt in contours[1:]:
            area = cv.contourArea(cnt)
            if area >= max_area[0]:
               max_area = [area, cnt]

    # Approximating the pupil with a circle
    # Finding its center coordinate (x, y) and its radius
        (x, y), radius = cv.minEnclosingCircle(max_area[1])
        x, y, radius = int(x), int(y), int(radius)
        cv.circle(self.image_matrix, (x, y), radius, (255, 153, 153), 2)
        return x, y, radius


# # driver code
# image = cv.imread("IR.jpg")
# PD = PupilDetector()
# (x, y, r), pupil = PD.detect(image)
# print("x:{}, y:{}, r:{}".format(x, y, r))
# cv.imshow("final", pupil)
# cv.waitKey(0)

# video driver code
# pupil video credit: https://youtu.be/vAgGeLJ37iU
cap = cv.VideoCapture('/content/ABA Therapy - Social Engagement.mp4')
PD = PupilDetector()
status = True
radius = None
font = cv.FONT_HERSHEY_SIMPLEX
org = (180, 220)
fontScale = 1
thickness = 2
while cap.isOpened():
    ret, frame = cap.read()
    # check whether there is any frame to read or not
    if ret:
        (x, y, r), pupil = PD.detect(frame)
        if (x, y, r) is not None:  # Check if (x, y, r) is not None
            if status:
                radius = r
                status = False
            if r is not None and radius is not None:  # Check if r and radius are not None
                if abs(r - radius) >= 10:
                    cv.putText(pupil, 'No Pupil', org, font, fontScale, (255, 255, 255), thickness, cv.LINE_AA)
                else:
                    cv.putText(pupil, 'Pupil', org, font, fontScale, (255, 255, 255), thickness, cv.LINE_AA)
            cv2_imshow( pupil)
    else:
        break
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

```python
"""
Demonstration of the GazeTracking library.
Check the README.md for complete documentation.
"""

import cv2
from gaze_tracking import GazeTracking

gaze = GazeTracking()
webcam = cv2.VideoCapture(0)

while True:
    # We get a new frame from the webcam
    _, frame = webcam.read()

    # We send this frame to GazeTracking to analyze it
    gaze.refresh(frame)

    frame = gaze.annotated_frame()
    text = ""

    if gaze.is_blinking():
        text = "Blinking"
    elif gaze.is_right():
        text = "Looking right"
    elif gaze.is_left():
        text = "Looking left"
    elif gaze.is_center():
        text = "Looking center"

    cv2.putText(frame, text, (90, 60), cv2.FONT_HERSHEY_DUPLEX, 1.6, (147, 58, 31), 2)

    left_pupil = gaze.pupil_left_coords()
    right_pupil = gaze.pupil_right_coords()
    cv2.putText(frame, "Left pupil:  " + str(left_pupil), (90, 130), cv2.FONT_HERSHEY_DUPLEX, 0.9, (147, 58, 31), 1)
    cv2.putText(frame, "Right pupil: " + str(right_pupil), (90, 165), cv2.FONT_HERSHEY_DUPLEX, 0.9, (147, 58, 31), 1)

    cv2.imshow("Demo", frame)

    if cv2.waitKey(1) == 27:
        break

webcam.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import tensorflow as tf
video_path = "/content/ABA Therapy - Social Engagement.mp4"
output_folder = "/content/extracted_frames"
# Function to preprocess video frames
def preprocess_frame(frame):
    resized_frame = cv2.resize(frame, (224, 224))  # Resize frame to match model input size
    normalized_frame = resized_frame / 255.0  # Normalize pixel values
    return normalized_frame
import os
import cv2
import numpy as np
import math
def extract_frames(video_path, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    video_name = os.path.basename(video_path)
    video_name_without_extension = os.path.splitext(video_name)[0]

    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    frame=[]
    for frame_num in range(frame_count):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_num)

        ret, frame = cap.read()

        if not ret:
            break

        x = frame_num % math.floor(frame_width / 200) * 200
        y = frame_num % math.floor(frame_height / 200) * 200

        cropped_frame = frame[y:y+200, x:x+200]

        output_image_path = os.path.join(output_folder, f"{video_name_without_extension}_frame_{frame_num}.jpg")
        cv2.imwrite(output_image_path, cropped_frame)
    return frame
frames = extract_frames(video_path, output_folder)

# Function to create a simple interaction detection model
def interaction_detection_model(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(64, activation='relu'),
        tf.keras.layers.Dense(2, activation='softmax')  # Output layer for binary interaction classification
    ])
    return model

# Load pre-trained interaction detection model (placeholder function)
def load_interaction_detection_model():
    # Placeholder function for loading pre-trained model
    # For simplicity, we'll create a new model here instead
    input_shape = (224, 224, 3)  # Input shape for the model
    model = interaction_detection_model(input_shape)
    return model
from keras.applications.vgg16 import preprocess_input
from keras.applications import VGG16

# Load pre-trained VGG16 model
vgg16 = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Function to extract features using VGG16
def extract_features(frame):
    # Resize frame to match VGG16 input size (224x224)
    frame = cv2.resize(frame, (224, 224))
    # Preprocess frame for VGG16 model
    frame = preprocess_input(frame)
    # Extract features using VGG16 model
    features = vgg16.predict(np.expand_dims(frame, axis=0))
    return features

# Function to detect interactions in a video
def detect_interactions(video_path):
    # Load pre-trained interaction detection model
    model = load_interaction_detection_model()

    # Open video capture
    cap = cv2.VideoCapture(video_path)

    # Loop through frames
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess frame
        preprocessed_frame = preprocess_frame(frame)

        # Extract features from frame
        features = extract_features(preprocessed_frame)

        # Predict interaction using the model
        interaction_probabilities = model.predict(np.expand_dims(features, axis=0))[0]

        # Assuming binary interaction classification
        interaction_label = np.argmax(interaction_probabilities)

        # Print or store interaction label

        # Example: Print interaction label
        print("Interaction:", interaction_label)

        # Example: Display frame with detected interaction
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break

    # Release video capture and close windows
    cap.release()
    cv2.destroyAllWindows()

# Example usage
detect_interactions('/content/ABA Therapy - Social Engagement.mp4')
```
